## Description du problème

Un pianiste souhaite jouer une mélodie donnée par une partition. La mélodie est une suite de n notes *(a0, a1, ... , an-1)* où chaque ai appartient à un ensemble de m notes numérotées de 0 à m-1.

Le pianiste peut choisir de jouer chaque note avec l’un de ses D doigts. Le passage de la note i1 jouée avec le doigt d1 à la note i2 jouée avec le doigt d2 est pénalisée par une fonction de difficulté de la transition *C(i1, d1, i2, d2)*. Notre pianiste souhaite donc déterminer avec quel doigt il doit jouer chaque note de la mélodie de façon à minimiser la somme des coûts de chaque transition.

Pour la partie empirique de ce TP, nous nous limiterons à des mélodies jouées à une main (D=5) sur deux octaves (m=24).

```python
def visualize_solution(costs, partition, solution):
    """
    Visualisez votre solution et les coûts associés.    
    """
    # Create a figure and axis
    _, ax = plt.subplots()
    
    # Plot the notes as circles with their sizes proportional to the partition values
    for i in range(len(partition)):
        ax.scatter([i], [partition[i]], marker='o')
        ax.text(i, partition[i] + 0.2, str(solution[i]), ha='center', va='bottom')
    
    # Set the limits for the axes to fit the plot
    ax.set_ylim(min(partition) - 1, max(partition) + 1)
    ax.set_xlim(-0.5, len(partition)-0.5)
    
    # Label the axes
    ax.set_xlabel('Partition')
    ax.set_ylabel('Note')
    ax.yaxis.set_major_locator(MaxNLocator(integer=True))
    
    # Title the plot
    ax.set_title('Partition et doigts utilisés')

    plot_costs = getCosts(costs, partition, solution).tolist()
    plot_costs.append(0)

    ax2 = ax.twinx()
    ax2.bar(range(len(partition)), plot_costs, width=1, alpha=0.5, align='edge', color='gray')
    ax2.set_xlim(-0.5, len(partition)-0.5)
    ax2.set_ylim(0, 100)
    
    # Remove the ticks and tick labels from the secondary y-axis
    ax2.tick_params(axis='both', which='both', length=0)
    ax2.set_yticklabels([])
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.spines['bottom'].set_visible(False)
    ax2.spines['left'].set_visible(False)    
    # Show the plot
    plt.show()

def generate_sample(size, maxValue=24):
    return np.random.randint(0, maxValue, size).tolist()

def get_dataset(seed, sizes):
    np.random.seed(seed)
    return { size: [generate_sample(size) for _ in range(5)]
        for size in sizes }

def readFile(filePath):
    """
    Charger une partition depuis un fichier
    """
    f = open(filePath)
    N = int(f.readline())
    partition = list(map(int, f.readline().split(' ')))
    f.close()
    return partition

def getCosts(costs, partition, solution):
    """
    Liste des coûts associés à chaque transition de votre solution.
    """
    prevNotes, nextNotes = partition[:-1], partition[1:]
    prevFingers, nextFingers = solution[:-1], solution[1:]
    return costs[prevNotes, prevFingers, nextNotes, nextFingers]

def getTotalCost(costs, partition, solution):
    """
    Coût total de votre solution
    """
    return np.sum(getCosts(costs, partition, solution))
```

```python
def measure(procedure, sample):
    """ Mesure le temps d'execution d'une procédure sur un unique exemplaire """
    start = time.time()
    solution = procedure(sample)
    end = time.time()
    sorting_time = end - start
    return solution, sorting_time


def measure_mean_time(procedure, samples):
    total_time = 0
    for i in range(len(samples)):
        sample = samples[i]
        _, compute_time = measure(procedure, sample)
        total_time += compute_time
        print(f"Computing {procedure.__name__} on sample {i+1} of size {len(sample)}...          ", end='\r')
    return total_time / len(samples)


def measure_procedure(procedure, dataset):
    """ Mesure les temps moyens d'execution d'une procédure sur chaque point d'un dataset """
    return {size: measure_mean_time(procedure, samples) for size, samples in dataset.items()}

def display_measures_table(measures):
    print("{: <12} {: <12}".format("Taille", "Temps moyen (s)"))
    for size, measure in measures.items():
        print("{: <12} {: <12}".format(size, measure))


def display_test_puissance(vals, title="Test de puissance"):
    x = list(vals.keys())
    y = list(vals.values())

    # Perform linear regression
    m, b, rvalue, _, _ = linregress(x, y)
    r = list(map(lambda x : m*x + b, x))

    # Create the line equation
    line_eq = f"y = {m:.2f}x + {b:.2f}"

    # Plot the points
    plt.scatter(x, y, label='Mesures')

    # Plot the regression line
    plt.plot(x, r, color="red", label=f'Regression linéaire R²={round(rvalue**2,4)}')

    # Add labels and title
    plt.xlabel('log Taille')
    plt.ylabel('log Temps')
    plt.title(title)

    # Add legend
    plt.legend(bbox_to_anchor=(0.60, 0), loc='lower left')

    # Display the line equation
    plt.text(min(x), max(y), line_eq)

    # Show the plot
    plt.show()

def display_test_rapport(vals, title="Test du rapport"):
    x = list(vals.keys())
    y = list(vals.values())

    plt.plot(x, y, label='Mesures')
    plt.scatter(x, y, label='Mesures')

    # Add labels and title
    plt.xlabel('Taille')
    plt.ylabel('Temps / f(taille)')
    plt.title(title)
    plt.show()


def display_test_constantes(vals, title="Test des constantes"):
    x = list(vals.keys())
    y = list(vals.values())

    # Perform linear regression
    m, b, rvalue, _, _ = linregress(x, y)

    r = list(map(lambda x : m*x + b, x))

    # Create the line equation
    line_eq = f"y = {m:.2E}x + {b:.2E}"

    # Plot the points
    plt.scatter(x, y, label='Mesures')

    # Plot the regression line
    plt.plot(x, r, color="red", label=f'Regression linéaire R²={round(rvalue**2,4)}')

    # Add labels and title
    plt.xlabel('f(Taille)')
    plt.ylabel('Temps')
    plt.title(title)

    # Add legend
    plt.legend(bbox_to_anchor=(0.60, 0), loc='lower left')

    # Display the line equation
    plt.text(min(x), max(y), line_eq)

    # Show the plot
    plt.show()
```

## Partie 1 : Algorithme glouton

Implantez un algorithme suivant le patron de conception glouton.

L'algorithme glouton fait son choix en fonction du coût de la prochaine transition :

Les deux premiers doigts d0 et d1 sont choisis de façon à minimiser C(a0, d0, a1, d1). Puis, étant donné le
doigt précédent d et la transition de la note i vers la note j, le choix glouton porte sur le doigt d’ qui
minimise C(i, d, j, d’).

Tentez de rendre votre implantation la plus performante possible en évitant des calculs inutiles.

```python
def glouton(partition):
    solution = [0 for _ in range(len(partition))]
    nb_digits = 5

    # Find the first 2 digits to minimize the cost of C(a0, d0, a1, d1)
    min_cost0 = float('inf')
    for digit0 in range(nb_digits):
        for digit1 in range(nb_digits):
            cost = costs[partition[0], digit0, partition[1], digit1]
            if cost < min_cost0:
                min_cost0 = cost
                solution[0] = digit0
                solution[1] = digit1

    for i in range(2, len(partition)):
        next_min_cost = float('inf')
        for next_digit in range(nb_digits):
            cost = costs[partition[i - 1], solution[i - 1], partition[i], next_digit]
            if cost < next_min_cost:
                next_min_cost = cost
                solution[i] = next_digit

    return solution

solutionGlouton = glouton(partition)
print("Coût Glouton :", getTotalCost(costs, partition, solutionGlouton))
visualize_solution(costs, partition, solutionGlouton)
```

### Analyse asymptotique

Dans cette analyse, vous tiendrez compte à la fois de la longueur de la mélodie mais également du
nombre de doigts pour chacun des trois algorithmes.

Pour *n* le nombre de notes (la longueur de la partition), et *D* le nombre de doigts.

L'utilisation mémoire est de l'ordre de *Theta(n)* pour la liste de notes.

L'algorithme commence par trouver le premier doigt, dans deux boucles imbriqués ayant autant d'itérations qu'il y a de doigts,
donc cela a une complexité de *Theta(D^2)*, qui est polynomial.

Le boucle suivant itèrera pour *(n - 2)* itérations en ayant une boucle interne de *D* itérations.
Donc, la complexité asymptotique de cet section est de *Theta(Dn)* en étant pseudo-polynomial.

Alors, l'algorithme a une complexité asymptotique de *Theta(max(D^2, Dn))*, pour être polynomial ou pseudo-polynomial dépendamment de si *D > n*.
Quoi que en pratique, *D < n*, puisqu'il y aura beaucoup plus de notes que de doigts.

Ainsi, l'algorithme est rapide et retourne une solution, mais il est peu probable que la solution soit optimale.
Et, dans notre cas, *D = 5* est constant, donc, pour l'analyse pratique, on s'attend à voir une complexité asymptotique de *Theta(n)* pour nos mesures uniquement.

## Partie 2 : Algorithme de programmation dynamique

L'algorithme de programmation dynamique remplit un tableau des coûts de transition optimaux en
fonction de la mélodie.

Notons `J(k, d)` la somme minimale des coût de transition associée à la mélodie `(ak, ... , an−1)` (c’est-à-
dire au suffixe de longueur `n - k` de la mélodie de départ) où l’on joue la première note ak avec le doigt
d.

Alors, J vérifie la relation de récurrence (1) :

J(k, d) = min { 0≤ d’ ≤ D-1} [ C(ak, d, ak+1, d’ ) + J(k + 1, d’ ) ]

et la condition terminale (2) :

J(n - 1, d) = 0, ∀d = 0, ... , D - 1

Comme on ne connaît pas le premier doigt à utiliser a priori, la solution au problème original est
donnée par (3) :

min { 0 ≤ d ≤ D-1} J(0, d)

Cette récurrence est résolue par programmation dynamique pour déterminer la solution au problème.

Implantez l'algorithme de programmation dynamique.

```python
def progDyn(partition):
    nb_digits = 5
    # shortest_solution[i], 'i' is the index of the partition note
    # shortest_solution[i] gives an array [] whose index are the digits 
    # the inner array contains tuples (next_best_digit, total_cost_from_the_end_to_here)
    shortest_solutions = [[(-1, 0) for _ in range(nb_digits)] for _ in range(len(partition))]
    
    # Calculate all path costs
    for i in range(len(partition) - 1, 0, -1):
        # We calculate the previous digit by comparing it to all the other next digits
        for prev_digit in range(nb_digits):
            min_cost = float('inf')
            for next_digit in range(nb_digits):
                transition_cost = costs[partition[i - 1], prev_digit, partition[i], next_digit]
                total_cost = transition_cost + shortest_solutions[i][next_digit][1]
                if total_cost < min_cost:
                    min_cost = total_cost
                    shortest_solutions[i - 1][prev_digit] = (next_digit, total_cost)

    # form solution
    solution = [0 for _ in range(len(partition))]

    # Get the best starting digit
    min_path_cost = float('inf')
    for digit in range(nb_digits):
        if shortest_solutions[0][digit][1] < min_path_cost:
            min_path_cost = shortest_solutions[0][digit][1]
            solution[0] = digit
    
    # Follow the rest of the following digits
    for i in range(0, len(partition) - 1):
        prev_digit = solution[i]
        # Set the next solution_digit based on the min_path of the previous digit
        solution[i + 1] = shortest_solutions[i][prev_digit][0]

    return solution

solutionDyn = progDyn(partition)
print("Coût Programmation Dynamique :", getTotalCost(costs, partition, solutionDyn))
visualize_solution(costs, partition, solutionDyn)
```

L'algorithme consiste de trois boucles imbriquées, la première boucle itère *n - 1* fois pour chacune des transitions de notes. La deuxième et la troisième boucle itère chacune *D* fois pour chacun des doigts et le contenu des boucles est une suite d'instructions de complexité asymptotique *Theta(1)*. ## Partie 3 : Algorithme de recherche locale

Cet algorithme explore le voisinage de solutions non optimales. En partant d’une solution gloutonne, la
solution est améliorée en tentant d’échanger un doigt pour un autre sur une note. Vous pourrez prendre
comme critère d’arrêt un nombre maximal d’itérations.

```python
def localSearch(partition):
    nb_digits = 5

    # Returns the sum of the two transitions costs from changing a current_digit
    # for the partition 0 and n only a single transition is considered
    def get_local_cost(index, current_digit):
        cost = 0
        if index > 0:
            cost += costs[partition[index - 1], solution[index - 1], partition[index], current_digit]
        if index < len(partition) - 1:
            cost += costs[partition[index], current_digit, partition[index + 1], solution[index + 1]]
        return cost


    # Get a base solution from the greedy algorithm
    solution = glouton(partition)
    solution_costs = [get_local_cost(i, solution[i]) for i in range(len(solution))]

    max_iterations = 3 * len(partition)
    for iteration in range(max_iterations):
        # get partition index from the iteration loop
        index = iteration % len(partition)

        for digit in range(nb_digits):
            digit_cost = get_local_cost(index, digit)

            if digit_cost < solution_costs[index]:
                solution[index] = digit
                solution_costs[index] = digit_cost

    return solution

solutionLocal = localSearch(partition)
print("Coût Recherche Locale :", getTotalCost(costs, partition, solutionLocal))
visualize_solution(costs, partition, solutionLocal)
```

### Analyse asymptotique

Dans cette analyse, vous tiendrez compte à la fois de la longueur de la mélodie mais également du
nombre de doigts pour chacun des trois algorithmes.

L'algorithme utilise uniquement une complexité asymptotique de mémoire de *Theta(n)* pour la taille de l'array de la solution.


Pour l'analyse asymptotique du temps d'exéctuon.
L'algorithme débute en formant une solution à partir de notre algorithme glouton que nous avons déterminé à avoir une complexité asymptotique de *Theta(max(D^2, Dn))*.
Puis on itère sur chacune des notes pour trouver le coût où la note est incluse dans un temps de *Theta(n)*, puisque la fonction utilitaire *get_local_cost()* est comprise de multiples instructions ayant un temps de *Theta(1)*.
Ensuite, il y a la boucle d'amélioration locale qui itèrera autant de fois que l'on décide. Ici, selon une expérimentation pratique, itérér *3n* fois semble être suffisant pour atteindre le maximum local de la solution. Ce seuil pourrait être raffiné, mais ici cela nous donne un facteur constant pour les itérations de la boucle externe. La boucle interne itère *D* fois pour chacun des doigts et ne contient que des opérations en temps de *Theta(1)*, alors la boucle en son ensemble itère *3Dn* fois qui fonctionne en complexité asymptotique *Theta(Dn)*, qui est un temps pseudo-polynomial.
Donc, l'algorithme, en sa globalité, a une complexité de *Theta(max(D^2, Dn))* tout comme l'algorithme glouton.

Cependant, selon nos observations, obtient des solutions significativement meilleure, mais ne peut atteindre que des maximums locaux et non la solution optimale globale.
Et, typiquement, *Dn > D^2*, puisque typiquement, une partition a plus de notes que de doigts à utiliser.
